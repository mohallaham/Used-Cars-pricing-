```python
import pandas as pd
from scipy.stats import f_oneway

# Example: Load your DataFrame (replace with your actual data)
# df = pd.read_csv("your_file.csv")

# Example dummy data

# Set the numerical target variable
numerical_col = 'Price'

# Select only categorical columns (excluding the target)
categorical_cols = df.select_dtypes(include='object').columns

# Store results
results = []

for col in categorical_cols:
    try:
        groups = [group[numerical_col].values for name, group in df.groupby(col)]
        if len(groups) > 1:  # ANOVA requires at least 2 groups
            f_stat, p_value = f_oneway(*groups)
            results.append((col, f_stat, p_value))
    except Exception as e:
        print(f"Error with column {col}: {e}")

# Convert to DataFrame and sort by p-value
anova_df = pd.DataFrame(results, columns=['Categorical_Var', 'F_statistic', 'p_value'])
anova_df = anova_df.sort_values('p_value')

print(anova_df)
```

```python
# Load the dataset

# Count the occurrences of each transmission type
transmission_counts = df['Transmission'].value_counts()
colors = sns.color_palette('pastel')
# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%', startangle=140, colors =colors)
plt.title('Distribution of Transmission Types in Car Listings')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.savefig("Trans_distribution.png", dpi=300, bbox_inches="tight", transparent=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the plot style and font scaling
sns.set(style="whitegrid")
sns.set_context("talk", font_scale=1.3)  # Increase overall font size

# Count plot for fuel types
plt.figure(figsize=(10, 6))
ax = sns.countplot(
    data=df,
    x='Fuel',
    order=df['Fuel'].value_counts().index,
    palette='viridis'
)

# Customize labels and title with bigger, bolder fonts
plt.title('Distribution of Fuel Types', fontsize=18, fontweight='bold')
plt.xlabel('Fuel Type', fontsize=16, fontweight='bold')
plt.ylabel('Number of Cars', fontsize=16, fontweight='bold')

# Make tick labels bigger and rotate x-axis labels
plt.xticks(rotation=45, fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

# Improve layout and save
plt.tight_layout()
plt.savefig("Fuel_distribution.png", dpi=300, bbox_inches="tight", transparent=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Value counts for the 'Condition' column
condition_counts = df['Condition'].value_counts()
colors = sns.color_palette('pastel')
# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', startangle=140, colors =colors)
plt.title('Car Condition Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.savefig("Condition_distribution.png", dpi=300, bbox_inches="tight", transparent=False)
plt.show()
```

```python
# Count the number of cars per year
year_counts = df['Year'].value_counts().sort_index()

# Line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.title('Number of Cars by Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.grid(True)
plt.tight_layout()
plt.savefig("Yeaar_distribution.png", dpi=300, bbox_inches="tight", transparent=False)
plt.show()
```

```python
# Scatter plot between Year and Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Year', y='Price', alpha=0.6)
plt.title('Scatter Plot of Year vs Price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
